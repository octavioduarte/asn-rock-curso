## Histograma

O histograma é um tipo de gráfico voltado para variáveis quantitativas contínuas:

- No eixo X: Teremos os intervalos de valores (chamados de classes ou bins).
- No eixo Y: A frequência (quantas observações cabem em cada intervalo).
