## Gráfico de Barras

- É um gráfico representado por barras.
- Cada barra representa a frequência ou outra medida de interesse analisada de cada categoria.
- Normalmente mais usado para dados qualitativos nominais.
- Para evitar interpretações erradas, devemos evitar o uso do 3D.


No gráfico de barras temos no eixo X (eixo horizontal) as variações dos valores qualitativos analisados e no eixo Y a quantidade de vezes em que este valor aparece no conjunto.

### Diferenças entre Gráfico de Barras e Histogramas


- Tipo de Dado: O gráfico de barras é usado para dados qualitativos (categóricos). Já o histograma é exclusivo para dados quantitativos contínuos (ex: altura, peso, temperatura), onde os valores são numéricos e fazem sentido em uma escala.

- Espaçamento entre as Barras: No gráfico de barras, há um espaço entre as barras porque cada uma representa uma categoria. No histograma, as barras são juntas, pois representam intervalos (determinados pelos beans) de uma variável contínua, mostrando a distribuição dos dados ao longo de um continuum.

- Objetivo da Análise: O gráfico de barras foca em comparar o tamanho de diferentes categorias. O histograma foca em visualizar a distribuição dos dados, identificando padrões como concentração, assimetria ou a forma da distribuição (ex: normal, exponencial).
