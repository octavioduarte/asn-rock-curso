## Boxplot

Boxplot é um tipo de gráfico utilizado para analisarmos a distribuição de um conjunto de dados numéricos. Ele "resume" essa distribuição focando em 6 valores:

- Valor mínimo antes da regra Q1 - 1.5IQR ("bigode" inferior)
- Q1 (P25)
- Q2 (P50 ou Mediana)
- Q3 (P75)
- Valor máximo antes da regra Q3+1.5IQR ("bigode" superior)
- Outliers



#### Anatomia de um boxplot

```markdown
     Q1-1.5IQR   Q1   median  Q3   Q3+1.5IQR
                  |-----:-----|
  o      |--------|     :     |--------|    o  o
                  |-----:-----|
outliers          <----------->            outliers
                       IQR
```


<br />

##### A caixa


A caixa representa os quartis do conjunto analisado. Por tanto imaginando um boxplot na horizontal (pode ser representado na vertical também) - A caixa contém 3 linhas verticais da esquerda pra direita cada linha corresponde a um dos quartis. 


```markdown
                Q1     Q2     Q3   
                  |-----:-----|
                  |     :     |
                  |-----:-----|                                                
```

É importante se atentar também ao tamanho da caixa que na prática seria a diferença entre o Q3 e Q1 chamado de diferença interquatilica abreviado como IQR - Este valor será importante mais adiante:

```markdown
                Q1     Q2     Q3   
                  |-----:-----|
                  |     :     |
                  |-----:-----|
                  <----------->
                       IQR
```

<br />

##### Os "bigodes"

Os bigodes representam os valores máximos e mínimos do conjunto que **não são outliers**:

<br />


Para o bigode inferior será considerado outlier valores menores que: **Q1-1.5IQR** 


Para o bigode superior será considerado outlier valores maiores que: **Q3+1.5IQR**


Os "bigodes" são os valores extremos antes dessa regra:


**VA_MENOR** = Valor antes da regra Q1-1.5IQR

**VA_MAIOR** = Valor antes da regra Q3+1.5IQR

```markdown
                            Q1   Q2    Q3   
                VA_MENOR   |-----:-----|   VA_MAIOR
               |-----------|     :     |--------|    
                           |-----:-----|
                           <----------->   
                                IQR
```

<br />

##### Outliers

Outliers são os valores que vão além da regra seja do bigode inferior ou superior - Por convenção são representados através de bolinhas nos boxplots:


```markdown
                            Q1   Q2    Q3   
                VA_MENOR   |-----:-----|   VA_MAIOR
   o           |-----------|     :     |--------|        o o
                           |-----:-----|
                           <----------->   
                                IQR
```
