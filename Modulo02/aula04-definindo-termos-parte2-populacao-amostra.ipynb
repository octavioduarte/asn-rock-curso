## Conhecendo outros termos utilizados na área de ciência de dados:

</br>

<h4>- População</h4>

<b><span style="color: blue">É o conjunto completo dos elementos</span></b> (indivíduos, objetos, eventos e etc) que possuem característica em comum e que são de interesse para um estudo ou análise. Exemplos: 

 - Todas as compras de um e-commerce
 - Todos os funcionários de uma empresa
 - Todos os animais de uma floresta

<hr/>

<h4>- Amostra</h4>

Um subconjunto da população. Muito usada para se ter uma maior compreensão da população sem a complexidade/barreira que uma análise em maior escala poderia exigir. Seguindo os exemplos de população acima: 

 - Compras do e-commerce dos últimos 12 meses
 - Um conjunto de funcionários de cada departamento
 - Todos os animais de um pedaço da floresta.

**Importante**: A amostra deve ser uma "fatia" da população bem tirada, ou seja, respeitando o contexto de negócio estabelecido, exemplo: Compras de e-commerce dos últimos 12 meses, porém se estivermos falando de 2020 (ano da pandemia) faz sentido eu seguir uma análise de um período atípico? Se não, deve-se pensar em uma amostra que não traga vieses incorretos.

<hr/>

<h4>- Estatísticas amostrais</h4>

São o resultado de análises feitos em uma amostra coletada para se ter inferência sobre a população, por tanto, coletamos a amostra de uma população nesta amostra podemos fazer análises que vão nos levar a uma compreensão da população. Levando em consideração que diferentes amostras de uma população podem (e provavelmente vão) levar a diferentes valores, é necessário compreender que:

<i style="color: blue"><b>"Na estatística, um número não é apenas um número — é um intervalo."</b></i>

Isso significa que os resultados de análises estatísticas raramente são valores exatos. Em vez disso, trabalhamos com intervalos que estimam onde o valor verdadeiro provavelmente está.

Exemplo:

**Contexto**: Coletar dados sobre os moradores do bairro Limoeiro.

**População**: Todos os moradores do bairro.

**Amostras**:

Amostra 1: Moradores da Rua A
**Estatística amostral** (média de idade): 38 anos

Amostra 2: Moradores da Rua B
**Estatística amostral** (média de idade): 42 anos

As duas amostras resultaram em médias diferentes — 38 e 42 anos. Com base nisso, não podemos afirmar com precisão uma única média para todo o bairro. Em vez disso, estimamos que a média de idade da população do bairro Limoeiro esteja dentro de um intervalo, por exemplo, entre 35 e 45 anos.

<h4>- Parâmetros e Estatística</h4>

<br/>

<span style="color: blue">Parâmetros</span>: São informações sobre nossa população, ou seja, sobre todo o conjunto de elementos do nosso estudo.


<span style="color: blue">Estatística</span>: São informações sobre uma amostra, ou seja, um subconjunto da população.

Em resumo: Em ambos os casos estamos falando de informações sobre nossos objetos de estudo, mas um se refere ao nosso conjunto todo e outro a uma parte dele. E uma outra diferença são os simbolos que representam essas informações, alguns exemplos:

<br/>

| Informação       | Parâmetro (População)| Estatística (Amostra)|
|------------------|----------------------|----------------------|
| Média            | μ                    | x̄                    |
| Variância        | σ²                   | s²                   |
| Desvio padrão    | σ                    | s                    |


<br/>

<i style="color: blue; font-size: 11px"><b>Todas as informações relacionadas a parâmetros são representadas com letras gregas.</b></i>

<i style="color: blue; font-size: 11px"><b>Todas as informações relacionadas a estatística são representadas com letras do nosso alfabeto.</b></i>
